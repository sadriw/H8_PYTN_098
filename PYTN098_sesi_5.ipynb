lebih complex lagi, kita bisa minta:
- khusus tim Nuggets
- year_id di antara 2010-2015
- yang game_result W win saja
- game_location boleh A atau N

```python
kon1 = df['fran_id'] == "Nuggets"

kon2 = df['year_id'] >= 2010
kon3 = df['year_id'] <= 2015

kon4 = df["game_result"] == 'W'

kon5 = df['game_location'] == "A"
kon6 = df['game_location'] == "N"

df_filtered = df[kon1 & (kon2 & kon3) & kon4 & (kon5 | kon6)]
df_filtered
```

```python
df_filtered["fran_id"].value_counts()
```

```python
df_filtered["year_id"].value_counts()
```

```python
df_filtered["game_result"].value_counts()
```

```python
df_filtered["game_location"].value_counts()
```

```python
df_filtered = df[
    (df['fran_id'] == "Nuggets") &
    (df['year_id'] >= 2010) &
    (df['year_id'] <= 2015) &
    (df["game_result"] == 'W') &
    ((df['game_location'] == "A") | (df['game_location'] == "N"))
  ]
df_filtered
```
